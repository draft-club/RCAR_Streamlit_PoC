```python
def _get_table_meta_data(table_dict):
    columns = table_dict[0]
    num_rows, num_columns = len(table_dict), len(columns)
    meta_data = {"columns":columns, "num_row":num_rows, "num_columns": num_columns}
    return meta_data
```
